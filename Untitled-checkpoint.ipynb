```python
### trying with 2 clusters at first 
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
clust_labels=model.fit_predict(df_copie)
df_copie['h']=pd.DataFrame(clust_labels)
```

```python
plt.scatter(df_copie[df_copie['h']==0]["PAYMENTS"], df_copie[df_copie['h']==0]["PURCHASES"], c='pink')
plt.scatter(df_copie[df_copie['h']==1]["PAYMENTS"], df_copie[df_copie['h']==1]["PURCHASES"], c=["c"])
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df_copie, method="ward"))
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=2, random_state=0) 
kmeans.fit(df_copie)
df_copie['k']=pd.DataFrame(kmeans.labels_)
```

```python
plt.scatter(df_copie[df_copie['k']==0]["PAYMENTS"], df_copie[df_copie['k']==0]["PURCHASES"], c='pink')
plt.scatter(df_copie[df_copie['k']==1]["PAYMENTS"], df_copie[df_copie['k']==1]["PURCHASES"], c=["c"])
```

```python
# the clusters do not look well devided so i'll look for better K 
```

```python
sum_distance = []
K = range(1,15)
for k in K:
    kmeans=KMeans(n_clusters=k)  
    kmeans = kmeans.fit(df_copie)
    sum_distance.append(kmeans.inertia_)
```

```python
plt.plot(K,sum_distance,'bx-')
plt.show()
```

```python
#according to the elbow methos , the best k we can use is 3
```
